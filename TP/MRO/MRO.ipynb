```python
from pycsp3 import *


import json

with open("donnees_cop/celar_50_7_10_5_0.800000_0.json", "r") as f:
    data = json.load(f)


nbr_stations = len(data['stations'])
nbr_regions = len(data['regions'])
stations = data['stations']
regions = data['regions']
delta = data['delta']
delta_min = data['delta_min']
max_frequencies_per_region = data['max_frequencies_per_region']


frequencies = VarArray(size=nbr_stations, dom=range(1, 100))  # Exemple de domaine de fréquences

# Contraintes
satisfy(
    # Contraintes de distance entre les fréquences des stations
    [abs(frequencies[i] - frequencies[j]) >= delta_min[i][j] for i in range(nbr_stations) for j in range(i+1, nbr_stations)],

    # Contraintes de nombre de fréquences par région
    [Count(frequencies[stations_in_region], value) <= max_frequencies_per_region[region] 
     for region, stations_in_region in enumerate(regions)]
)

# Problèmes d'optimisation
minimize(
    # Minimiser le nombre de fréquences utilisées
    NValues(frequencies),

    # Utiliser les fréquences les plus basses possibles
    Sum(frequencies),

    # Minimiser la largeur de la bande de fréquences utilisées
    Max(frequencies) - Min(frequencies)
)

# Résoudre le problème
if solve():
    print("Solution trouvée :")
    print(frequencies)
else:
    print("Pas de solution trouvée.")
```
