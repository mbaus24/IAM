```python
def char(sentence):
    # Sparer les mots par le caractre <esp>
    char = ['<pad>']
    
    for word in sentence:
        
        if word == ' ':
            char.append('<esp>')
        else:
            char.append(word)
    # Ajouter un padding au dbut
    return char

def i_enc(chars):
    indices = []
    char2idx = {'<pad>': 0, '<unk>': 1, '<esp>': 2}
    for char in chars:
        if char not in char2idx:
            char2idx[char] = len(char2idx)
        indices.append(char2idx[char])
    return indices, char2idx




def ends(chars):
    ends = []
    end = 0
    for char in chars:

        if char == '<esp>':
            ends.append(end-1)
        end += 1
    return ends

# Example usage
phrase = "L'association a chang les dcors et avec l'aide de plusieurs bnvoles, tabli diffrents tableaux sur le thme de \"Cinq sicles d'activit conomique de la rgion d'Ancerville\"."
chars = char(phrase)
ends = ends(chars)
indices, char2idx = i_enc(chars)

# print(chars)
# print(ends)
# print(indices)

def preparation(sentence):
    chars = char(sentence)
    end = ends(chars)
    indices, char2idx = i_enc(chars)
    return chars, end, indices, char2idx

```

```python
import torch as nn

class GRUClassifier(nn.Module):
    def __init__(self, d_embed, d_hidden, d_in, d_out):
        super().__init__()
        self.embed = nn.Embedding(d_in, d_embed, padding_idx=0)
        self.gru = nn.GRU(d_embed, d_hidden, batch_first=True, bias=False)
        self.dropout = nn.Dropout(0.1)
        self.decision = nn.Linear(d_hidden, d_out)
    
    def forward(self, idx_words):
        embedded = self.embed(idx_words)
        hidden, _ = self.gru(embedded)
        hidden = hidden.contiguous().view(-1, hidden.size(-1))  # Flatten hidden states
        return self.decision(self.dropout(hidden))
```
