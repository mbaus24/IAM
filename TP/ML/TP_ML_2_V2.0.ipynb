```python
def lexique(data):
    f = open(data, "r", encoding="utf-8")
    lines = [[x.rstrip('\n')] for x in f]
    lexique = []
    comptage = []
    #print(lines)
    for phrase in lines:

        temp=phrase[0].split(" ")
        for mot in temp:
            if mot not in lexique:

                lexique.append(mot)
                comptage.append(1)
            else:
                comptage[lexique.index(mot)]+=1
    return lexique, comptage

```

```python
# END:
def enleve_mot(texte, m, count): #renvoie les mots correspondants au lexique filtré de texte sous forme de liste de mots (sans les comptages)
    if count==True:
          return [item for item in list(zip(lexique(texte)[0], lexique(texte)[1])) if item[1] >= m]
    else:
          return [item[0] for item in list(zip(lexique(texte)[0], lexique(texte)[1])) if item[1] >= m]

data = "data/La_Reine_Margot.tok"

filtered_results_count = enleve_mot(data, 3, True)
filtered_lex = enleve_mot(data, 3, False)

print(filtered_results_count)
print(len(filtered_results_count))
```

```python
data = "data/La_Reine_Margot.tok"
def strip_phrases(data): #pour un texte donné, renvoie une liste de ses phrases, où les phrases sont décomposées par liste de mots
  f = open(data, "r", encoding="utf-8")
  lines = [[x.rstrip('\n')] for x in f]
  phrases=[]
  for phrase in lines:
    temp=phrase[0].split(" ")
    phrases.append(temp)
  return phrases

strip_phrases(data)
```

```python
def contexte(m, L, txt):  #pour un mot m, renvoie les mots du contexte de m de taille L dans le texte filtré
  contexte_mot=[]                        #les mots qui n'appartiennent pas au lexique filtré ne sont pas ajoutés au contexte
  filtered_data=enleve_mot(txt, 3, False)
  for phrases in strip_phrases(data):
    if m in phrases:
      ind = phrases.index(m)
      for i in range(max(ind-L,0), ind):
        if (phrases[i] not in contexte_mot) and (phrases[i] in filtered_data):
          contexte_mot.append(phrases[i])
      for i in range(ind+1, min(ind+L+1,len(phrases))):
        if (phrases[i] not in contexte_mot) and (phrases[i] in filtered_data):
          contexte_mot.append(phrases[i])
  return contexte_mot

contexte("louvre", 1, data)
```
