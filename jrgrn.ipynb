```python
from collections import Counter

# Path to your input CSV file
bda = 'bda.csv'
bde = 'bde.csv'
bds = 'bds.csv'
bds2 = 'bds2.csv'
bda2 = 'bda1.csv'
bda3 = 'bda3.csv'
# Load the CSV file
data = pd.read_csv(bda3)

# Extract the fifth column (index 4 since Python indexing starts at 0)
fourth_column = data.iloc[:, 4]

# Initialize first_match as None
first_match_bda = []

# Iterate over the fourth column to find the first match
for line in fourth_column:
    matches = re.findall(pattern, line)
    if matches[0] != 'Blanc':
        first_match_bda.append(matches[0])

data = pd.read_csv(bds)

# Extract the fifth column (index 4 since Python indexing starts at 0)
fourth_column = data.iloc[:, 4]

# Initialize first_match as None
first_match_bds = []

# Iterate over the fourth column to find the first match
for line in fourth_column:
    
    matches = re.findall(pattern, line)
    if matches[0] !='Blanc':
        first_match_bds.append(matches[0])

data = pd.read_csv(bde)

# Extract the fifth column (index 4 since Python indexing starts at 0)
fourth_column = data.iloc[:, 4]

# Initialize first_match as None
first_match_bde = []

# Iterate over the fourth column to find the first match
for line in fourth_column:
    matches = re.findall(pattern, line)
    if matches[0] != 'Blanc':
        first_match_bde.append(matches[0])


plt.hist(first_match_bda, label='bda', alpha=0.5, align="mid", rwidth=1)  # density=False would make counts
plt.hist(first_match_bds, label='bds', alpha=0.5, align="mid", rwidth=1)  # density=False would make counts
plt.hist(first_match_bde, label='bde', alpha=0.5, align="mid", rwidth=1)  # density=False would make counts

plt.xlabel('Liste')
plt.ylabel('Repartition des votes')
plt.xticks(rotation=90)
plt.legend()
plt.show()

# Count the occurrences of each unique value in first_match_bds

bds_count = Counter(first_match_bds)
print(bds_count)
```
