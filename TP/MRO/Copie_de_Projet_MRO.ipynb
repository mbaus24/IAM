```python
import json

def generate_wcsp_file(data, filename):
    nbr_stations = len(data['stations'])
    tableau_freq = [[data['stations'][i]['emetteur'], data['stations'][i]['recepteur']] for i in range(nbr_stations)]
    delt = [data['stations'][i]['delta'] for i in range(nbr_stations)]
    delta = [[0 for i in range(nbr_stations)] for j in range(nbr_stations)]
    for i in range(len(data['interferences'])):
        x = data['interferences'][i]['x']
        y = data['interferences'][i]['y']
        delta[x][y] = data['interferences'][i]['Delta']

    liaison = [(data['liaisons'][i]['x'], data['liaisons'][i]['y']) for i in range(len(data['liaisons']))]


    num_variables = nbr_stations * 2

    max_domain_size = 0
    for station_freq in tableau_freq:
        max_domain_size = max(max_domain_size, len(station_freq[0]), len(station_freq[1]))
    max_cost = 1000  # Coût max à ajuster

    # Estimation du nombre de contraintes (sans les régions)
    num_constraints = nbr_stations + sum(delta[i][j] > 0 for i in range(nbr_stations) for j in range(nbr_stations)) + len(liaison)

    with open(filename, "w") as f:
        f.write(f"// Celar Frequency Assignment\n")
        f.write(f"{num_variables} {max_domain_size} {num_constraints} {max_cost}\n")

        # Variables
        for i in range(nbr_stations):
            for j in range(2):
                f.write(f"{i * 2 + j + 1} {len(tableau_freq[i][j])} {' '.join(map(str, tableau_freq[i][j]))}\n")

        # Contraintes delt
        for i in range(nbr_stations):
            f.write(f"2 {i * 2 + 1} {i * 2 + 2} 0 {len(tableau_freq[i][0]) * len(tableau_freq[i][1])}\n")
            for freq1 in tableau_freq[i][0]:
                for freq2 in tableau_freq[i][1]:
                    cost = 0 if abs(freq1 - freq2) == delt[i] else max_cost
                    f.write(f"{freq1} {freq2} {cost}\n")

        # Contrainte delta
        for i in range(nbr_stations):
            for j in range(nbr_stations):
                if delta[i][j] > 0:
                    f.write(f"4 {i * 2 + 1} {i * 2 + 2} {j * 2 + 1} {j * 2 + 2} 0 {len(tableau_freq[i][0]) * len(tableau_freq[i][1]) * len(tableau_freq[j][0]) * len(tableau_freq[j][1])}\n")
                    for freq1 in tableau_freq[i][0]:
                        for freq2 in tableau_freq[i][1]:
                            for freq3 in tableau_freq[j][0]:
                                for freq4 in tableau_freq[j][1]:
                                    cost = 0 if min(abs(freq1 - freq3), abs(freq1 - freq4), abs(freq2 - freq3), abs(freq2 - freq4)) >= delta[i][j] else 10
                                    f.write(f"{freq1} {freq2} {freq3} {freq4} {cost}\n")

        # Contraintes de liaison
        for i, j in liaison:
            f.write(f"2 {i * 2 + 1} {j * 2 + 2} 0 {len(tableau_freq[i][0]) * len(tableau_freq[j][1])}\n")
            for freq1 in tableau_freq[i][0]:
                for freq2 in tableau_freq[j][1]:
                    cost = 0 if freq1 == freq2 else 50
                    f.write(f"{freq1} {freq2} {cost}\n")
```
