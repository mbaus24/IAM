### `numpy` (`np`)

La bilbiothèque `numpy` (`np`) est une biliothèque Python utilisée pour faire des calculs d'algèbre linéaire. Elle propose un support pour une utilisation simplifiée et optimisée des vecteurs et matrices (tableaux) grâce à la classe `np.ndarray`. Nous allons à présent voir un exemple concret et extremement simple où l'utilisation de la classe `np.ndarray` nous fait gagner du temps par rapport à l'utilisation des listes: comment récupérer dans une variable une colonne en particulier dans notre tableau? C'est à dire que pour chaque ligne de la liste, on veut récupérer une valeur à une position donnée, disons la position 3. Nous verrons qu'il y a un gain de temps en terme de "calcul" mais aussi en terme d'écriture de code.

Avec les listes, nous ferions quelquechose comme:

```python
colonne_3 = []
for l in data_list:
    colonne_3.append(l[3])
print(colonne_3[:10])
```

En utilisant `np.ndarray`, nous devons d'abord transformer notre liste de listes en *array* `numpy`:

```python
data_arr = np.array(data_list)
print(data_arr)
print(type(data_arr))
```

```python
print (data_arr.shape)
```

Ensuite, pour récupérer la colonne 3 de cet *array*, on utilise les coupes d'indexation qui étendent le concept standard d'indexation des listes aux tableaux à N dimensions. Dans notre cas, nous avons un tableau en 2 dimensions (les lignes et les colonnes).

Soyons plus concrets: avec des listes standard, l'indexation se fait avec **_un_ entier et rien d'autre** (`data_list[n]`) alors qu'avec un `np.ndarray`, l'indexation peut se faire avec **un _tuple_ d'entiers** de taille < N (le nombre de dimensions) (`data_arr[a, b]`) où chaque entier du tuple faire référence aux éléments à récupérer sur la dimension correspondante.

Ainsi, pour récupérer une colonne en particulier (ici, la colonne 3), on fait:

```python
colonne_3 = data_arr[:, 3]
print(colonne_3[:10])
```

Notez l'utilisation des `:`: ils signifient "tous". Dans ce cas, ils veulent dire "toutes les lignes". Donc la première ligne de cette cellule signifie "Stocke dans colonne_3 l'élément 3 de chaque ligne de data_arr".

#### Digression: Détails sur l'indexation avec `numpy`

Ici, nous allons nous poser un petit peu pour comprendre ce qu'il se passe. C'est **très important** que vous compreniez l'indexation avec `numpy` pour éviter des erreurs dans le futur.

##### Indexation simple

Lorsque nous avons récupéré la **liste** `colonne_3`, nous avons crée une nouvelle liste que nous avons rempli avec la troisième valeur de chaque ligne du tableau `data_list`. Ces valeurs ont donc été **copiées** et sont donc présentes en deux endroits de la mémoire vive de l'ordinateur: dans `colonne_3` et dans `data_list`.

A l'inverse, avec le système de coupes de `numpy`, `colonne_3` est une **vue** de la troisième colonne de `data_arr`, c'est à dire que les données n'ont pas été copiées: `colonne_3` et `data_arr[:, 4]` sont deux références à la **même** portion de mémoire vive de l'ordinateur. Attention donc, **modifier `colonne_3` revient à modifier `data_arr[:, 4]`** et donc `data_arr` en général.

Pour vous en convaincre, voici un morceau de code où je crée une matrice de 1, j'en récupère une vue et je fais une modification sur cette vue qui se répercute sur la matrice initiale.

```python
ones = np.ones((3, 3))
print("matrice", ones)
col_2 = ones[:, 2]  # vue de la dernière colonne de la matrice
print("colonne", ones[:, 2])
print("vue colonne", col_2)

print("\n")
col_2[2] = 2  # modification de la dernière valeur de la vue
print("vue colonne", col_2)
print("colonne", ones[:, 2])
print("matrice", ones)
```

#### Indexation avancée

Puisque nous y sommes, faisons un tour vers les techniques d'indexation avancée offertes par `numpy`:

- l'indexation par tableau d'indices;
- l'indexation par tableau de booleens.

On utilise l'indexation par tableau d'indices pour spécifier précisemment quels éléments (plusieurs éléments) on veut récupérer en particulier en spécifiant explicitement les éléments que l'on veut:

```python
arr = np.random.rand(3, 3)  # cree une matrice 3x3 aléatoire
print(arr)
print()
sub_arr = arr[[0, 2]]  # récupérer la première et la dernière ligne
print(sub_arr)
```

On utilise l'indexation par tableau de booleens pour la même raison mais cette fois, on utilise un tableau de booleen de même taille que le tableau où on veut indexer. Les éléments à `True` sont retournés dans le sous-array indexé. Habituellement, ce tableau est obtenu à partir d'une condition sur le tableau initial:

```python
arr = np.random.rand(3)  # cree un vecteur de taille 3 aléatoire
print(arr)
bool_arr = arr > 0.25  # cree un vecteur de booleen de taille len(arr)
# dont chaque composante vaut True si la composante associée dans arr est > 0.25
print(bool_arr)
sub_arr = arr[bool_arr]
print(sub_arr)
```

**Attention**, cette fois, ces méthodes d'indexation avancées ne renvoies pas de vues mais des copies! Voyez plutôt:

```python
arr = np.ones(3)  # cree un vecteur de 1 de taille 3
print("array", arr)
sub_arr = arr[[0, 2]]  # récupérer le premier et dernier element
print("sub_array", sub_arr)
print()

sub_arr[0] = 2 # modifier le premier element du sub_array
print("sub_array", sub_arr)
print("arr", arr)  # l'array inital n'a pas été modifié!
```

Ce passage sur l'indexation des *array* `numpy` était un peu lourd mais je pense que c'était important de passer par là. Retenez *au moins* qu'avec l'indexation simple, on récupère une **vue** de l'*array* initial et non pas une **copie**

Parmi ces commandes magiques, vous trouverez `%time`, `%timeit`, `%%time` et `%%timeit`. La version avec un seul `%`  sert à timer des programmes d'une seule ligne alors qu'avec deux `%%`, on peut timer des programmes de plusieurs lignes. La difference entre `time` et `timeit` est plus subtile: `time` retourne le temps réel vécu par l'utilisateur alors que `timeit` renvoi le temps mis par le système pour exécuter exactement l'instruction demandée (en plus, `timeit` execute plusieurs fois l'instruction et renvoie la moyenne et l'écart type).

En gros, vous pouvez utiliser la méthode que vous voulez mais soyez pertinents et constants (toujours la même). Moi je préfère `timeit`. Voyez donc comment je time la sélection de la colonne 3 de notre tableau de données:

```python
%timeit colonne_3 = data_arr[:, 3]
```

```python
%%timeit

colonne_3 = []
for l in data_list:
    colonne_3.append(l[3])
```

La version avec `numpy` est, en gros, 1 million de fois plus rapide que la version avec la liste. Convaincus? ;)

Notez qu'ici, il n'y a aucune (ou presque) modification de la mémoire avec le mécanisme des vues de `numpy` alors qu'avec les listes, on écrit 150,000 nouvelles valeurs dans la mémoire vive. Donc c'est pas très équitable: pour timer équitablement la copie avec `numpy`, utilisons la méthode d'indexation avancée par tableau d'indices (notez que le 3 est dans une liste, cette fois-ci):

```python
%timeit colonne_3 = data_arr[:, [3]]
```

La version avec `numpy` est toujours plus rapide (3x) pour cette opération très simple. Imaginez avec des opérations plus complexes! En bref, quand vous manipulez beaucoup de données (numériques, notamment), utilisez `numpy` et surtout pas les listes (et évitez les boucles autant que possible).

##### `df.loc`

Les positions au sein d'un *dataframe* sont définies par un *index* et une colonne. Si vous affichez `df` dans Jupyter (vu plus haut), vous verrez sur la gauche des chiffres affichés en gras, la liste de ces chiffres constitue l'*index* du *dataframe*. En haut, vous trouvez les titres des colonnes du *dataframe*. La liste de ces colonnes constitue les colonnes du *dataframe*. Les attributs `df.index` et `df.columns` permettent d'afficher respectivement l'*index* et les colonnes du *dataframe*.

Pour sélectionner une ou des positions du *dataframe* en utilisant l'*index* et les colonnes, on utilise la fonction `df.loc` qui s'utilise comme un `np.ndarray` de deux dimensions mais qui supporte les chaines de caractère. La cellule suivante affiche le numéro de l'affaire, le sexe de l'accidenté, son age et son poids pour les accidentés 2 à 8.

```python
df.loc[2:8, ["CASENUM", "SEX_IM", "AGE_IM", "WEIGHT"]]
```

##### `df.iloc`

La deuxième méthode de sélection est encore plus proche des `np.ndarray` car elle ne fonctionne qu'avec des indices. On s'en sert de la même façon que `df.loc` mais au lieu de donner les noms explicites des éléments de `df.index` et `df.columns`, on utilise leurs indices. Ca ne fait pas de différence (dans notre cas!! mais pas toujours) pour `df.index`. Pour obtenir la même sortie que précédemment, on fait:

```python
df.iloc[2:8, [0, 3, 8, 10]]
```

##### Fusion/jointure

Pandas permet aussi de réaliser des jointures entre tableaux qui partagent des clés afin de gérer dans un seul tableau des données qui sont liées entre elles. Dans l'exemple suivant, nous simulons la présence de deux tableaux à partir de ntore *dataframe* initial puis nous réaliserons une jointure sur ceux-ci.

```python
drivers = df[df.PER_TYP==1]  # selection des lignes du dataframe qui concernent les passagers
passengers = df[df.PER_TYP==2]  # selection des lignes du dataframe qui concernent les conducteurs
passengers = passengers[["CASENUM", "VEH_NO", "PER_NO", "INJ_SEV"]]  # sous-selection de certaines colonnes des passagers
new = pd.merge(drivers, passengers, on=["CASENUM", "VEH_NO"], how="inner", copy=False)  # jointure
new
```

```python
new.shape
#df.shape
```

Détaillons cette jointure:
- *drivers* et *passengers*: les deux *dataframe* que nous fusionnons
- *on*: les clés utilisées pour comparer les lignes des deux dataframe
- *how*: la méthode de jointure utilisée. Ici *inner* signifie que nous allons joindre les lignes des deux dataframe dont les *CASENUM* et *VEH_NO* sont identiques dans les deux dataframe. Les autres lignes sont supprimées du résultat
- *copy*: permet de ne pas faire de copies inutiles. Ne mord pas. (je ne sais pas exactement ce que ça veut dire)
