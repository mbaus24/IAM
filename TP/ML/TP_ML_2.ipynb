```python
def sigmoide(x):
  return (1/(1+np.exp(-x)))

dim=5
k=2


m = torch.rand((dim), requires_grad=True)
cpos = torch.rand((dim), requires_grad=True)
cneg = torch.rand((k,dim), requires_grad=True)

loss = -(torch.log(torch.nn.functional.sigmoid(cpos.T @ m)) + torch.sum(torch.log(torch.nn.functional.sigmoid(-cneg @ m))))

loss.backward()

print(m.grad, cpos.grad, cneg.grad)

def gradient(cpos, cneg, m):
  gradientcpos = (sigmoide(np.dot(m,cpos))-1)*m
  gradientcneg = (sigmoide(np.dot(m,cneg)*m))
  gradientm = (sigmoide(np.dot(m,cpos))-1)*cpos + np.sum((sigmoide(np.dot(m,cneg))*cneg))

  return gradientcpos, gradientcneg, gradientm


def majplongement(cpos, cneg, m, eta):
    cpos=cpos-eta*gradient(cpos,cneg,m,eta)[0]
    cneg=cneg-eta*gradient(cpos,cneg,m,eta)[1]
    m=m-eta*(gradient(cpos,cneg,m,eta)[2])
    return cpos, cneg, m

data = "data/La_Reine_Margot.tok"
def strip_phrases(data): #pour un texte donné, renvoie une liste de ses phrases, où les phrases sont décomposées par liste de mots
  f = open(data, "r", encoding="utf-8")
  lines = [[x.rstrip('\n')] for x in f]
  phrases=[]
  for phrase in lines:
    temp=phrase[0].split(" ")
    phrases.append(temp)
  return phrases

strip_phrases(data)

filtered_lex=enleve_mot(data, 3, False)


def contexte(m, L, filtered_data):  #pour un mot m, renvoie les mots du contexte de m de taille L dans le texte filtré
  contexte_mot=[]                   #les mots qui n'appartiennent pas au lexique filtré ne sont pas ajoutés au contexte
  for phrases in strip_phrases(data):
    if m in phrases:
      ind = phrases.index(m)
      for i in range(max(ind-L,0), ind):
        if (phrases[i] not in contexte_mot) and (phrases[i] in filtered_data):
          contexte_mot.append(phrases[i])
      for i in range(ind+1, min(ind+L+1,len(phrases))):
        if (phrases[i] not in contexte_mot) and (phrases[i] in filtered_data):
          contexte_mot.append(phrases[i])
  return contexte_mot

contexte("louvre", 3, filtered_lex)

```
