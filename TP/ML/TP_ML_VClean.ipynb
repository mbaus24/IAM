```python
def lexique(data):
    f = open(data, "r", encoding="utf-8")
    lines = [[x.rstrip('\n')] for x in f]
    lexique = []
    comptage = []
    #print(lines)
    for phrase in lines:

        temp=phrase[0].split(" ")
        for mot in temp:
            if mot not in lexique:

                lexique.append(mot)
                comptage.append(1)
            else:
                comptage[lexique.index(mot)]+=1
    return lexique, comptage

```

```python
#ici, txt sera le lexique filtré avec comptage
#implémentation de la méthode du rejet, pour ne pas acoorder trop de poids aux mots rares en choisissant aléatoirement
#les mots choisis ne doivent pas être inclus dans liste_interdite, qui sera ici notre contexte d'un mot
def rejet(txt, taille, liste_interdite, tot_mots):

  L=[]
  while len(L) < taille:
    k = random.randint(0,len(txt)-1)
    #print(txt[k][1])
    proba = (txt[k][1])/(tot_mots)
    i = random.uniform(0,1)
    if (i<proba) and (txt[k][0] not in liste_interdite) and (txt[k][0] not in L):
      L.append(txt[k][0])
  return L

#rejet(filtered_lex_count, 20, contexte('louvre', 2, filtered_lex), tot_mots_lex)
```

```python
def select_cposcneg(m, L, filtered_txt, filtered_txt_count, k, tot_mots_lex):
  Mots=[]
  cont=contexte(m, L, filtered_txt)
  a = cont[random.randint(0,len(cont)-1)]
  inda = filtered_txt.index(a)
  Mots.append(inda)
  rej = rejet(filtered_txt_count, k, cont, tot_mots_lex)
  Mots_neg=[]
  for i in range(0, k):
      b = rej[i]
      indb = filtered_txt.index(b)
      Mots_neg.append(indb)
  Mots.append(Mots_neg)
  return Mots

#select_cposcneg('louvre', 2, filtered_lex, filtered_lex_count, 10, tot_mots_lex)
```

```python
import torch


def w2v(txt, n, L, k, eta, e, minc):

  if torch.cuda.is_available():
    device = torch.device('cuda')
    print('GPU available')
  else: 
    device = torch.device('cpu')

  filtered_lex_count = enleve_mot(txt, minc, True)
  filtered_lex = enleve_mot(txt, minc, False)

  tot_mots_lex = 0
  for i in range(len(filtered_lex_count)):
    tot_mots_lex+=filtered_lex_count[i][1]
  tot_mots_lex

  M=torch.randn((len(filtered_lex), n), requires_grad=True, device=device)
  C=torch.randn((len(filtered_lex), n), requires_grad=True, device=device)

  for iter in range(e):
    print('Itération '+str(iter+1))
    for mot in range(len(filtered_lex)):
      #print(filtered_lex[mot])
      batch = select_cposcneg(filtered_lex[mot], L, filtered_lex, filtered_lex_count, k, tot_mots_lex)
      mot_plong = M[mot]
      cpos_plong = C[batch[0]]

      cneg_plong = torch.empty((0, n), requires_grad=True, device=device)
      for cneg in batch[1]:
        a = torch.reshape(C[cneg], (1,n))
        cneg_plong = torch.concat((cneg_plong, a), 0)

      mot_plong.retain_grad()
      cpos_plong.retain_grad()
      cneg_plong.retain_grad()

      loss = -(torch.log(torch.nn.functional.sigmoid(cpos_plong.T @ mot_plong)) + torch.sum(torch.log(torch.nn.functional.sigmoid(-cneg_plong @ mot_plong))))

      loss.backward(retain_graph=True)

      mot_plong = mot_plong - eta*(mot_plong.grad)
      cpos_plong = cpos_plong - eta*(cpos_plong.grad)
      cneg_plong = cneg_plong - eta*(cneg_plong.grad)

      with torch.no_grad():
        M[mot] = mot_plong
        C[batch[0]] = cpos_plong
        for i in range(0,len(batch[1])):
            C[batch[1][i]] = cneg_plong[i]

      '''mot_plong.grad.data.zero_()
      cpos_plong.grad.data.zero_()
      cneg_plong.grad.data.zero_()
  '''
  #print(M,C)
  plongements = [str(len(filtered_lex))+' '+ str(n)]

  with torch.no_grad():

    for i in range(len(filtered_lex)):
      delimiter = " " # Define a delimiter
      if torch.cuda.is_available():
        num_list_string = map(str, M.cpu()[i].numpy()) # Convert each element into a string first
      else:
        num_list_string = map(str, M[i].numpy())
      join_num_str = delimiter.join(num_list_string)
      plongements.append(filtered_lex[i] + ' ' + join_num_str)

  with open('plongements_' + 'results' +'.txt', 'w', encoding="utf-8") as f:
    for line in plongements:
        f.write(f"{line}\n")

  return M, C, plongements
```

# 1 Objectif
L'objectif de ce projet est de programmer un modèle de plongement de mots statique inspiré du modèle word2vec, plus spécifiquement de la méthode skip-gram with negative sampling.
Le plongement d'un mot m est un vecteur de réels m qui représente de manière abstraite les contextes possibles d'occurrence de m. Deux mots m et m' véhiculant le même sens, tel que les mots vélo et bicyclette, peuvent apparaître dans les mêmes contextes. Leurs plongements doivent alors être très proches.

# 2. Evaluation

```python
def cos_sim(a,b):
  return (a @ b.T)/(torch.norm(a)*torch.norm(b))
```

```python
def eval_w2v(plongements, eval):

  eval_txt = open(eval, "r", encoding="utf-8")
  plongement_txt = list(open(plongements, "r", encoding="utf-8"))

  nb_succes=0
  nb_echec=0

  for line in eval_txt:
    mot_eval = str(line.split(' ')[0])
    mot_pos = str(line.split(' ')[1])
    mot_neg = str(line.split(' ')[2][:-1])
    for plongement in plongement_txt:
      if plongement.split(' ')[0] == mot_eval:
        plong_eval = [float(i) for i in (plongement.split(' ')[1:])]
        tens_eval= torch.Tensor(plong_eval)
      elif plongement.split(' ')[0] == mot_pos:
        plong_pos = [float(i) for i in (plongement.split(' ')[1:])]
        tens_pos= torch.Tensor(plong_pos)
      elif plongement.split(' ')[0] == mot_neg:
        plong_neg = [float(i) for i in (plongement.split(' ')[1:])]
        tens_neg= torch.Tensor(plong_neg)

    #print (cos_sim(tens_eval, tens_pos), cos_sim(tens_eval, tens_neg))

    if cos_sim(tens_eval, tens_pos) > cos_sim(tens_eval, tens_neg):
      nb_succes += 1
    else:
      nb_echec += 1

  return nb_succes/(nb_succes + nb_echec)
```

##Fasttext

La méthode de construction de plongements implémentée ne permet pas de produire un plongement pour un mot qui n’a pas été vu dans le corpus d’apprentissage. Lorsque deux mots sont morphologiquement liés, comme mangions et mangerions, on pourrait essayer de construire le plongement de l’un à partir de l’autre. L’algorithme Fasttext propose une méthode pour faire cela en décomposant les mots en séquences de n-lettres, mangerions, par exemple, deviendrait " ma man ang nge ger eri rio ion ons ns " , si l’on choisit une décomposition en trigrammes. Pour chacun des trigrammes, un plongement est calculé et le plongement d’un mot est obtenu en faisant la somme des plongements des trigrammes qui le composent. Après avoir lu l’article de fasttext Enriching Word Vectors with Subword Information, implémentez la méthode proposée et proposez une manière intelligente de l’évaluer.


```python
#
def separation(data):
  '''cet algorithme va séparer le texte d'entrée en une liste de phrases,
    où chaque phrase est une liste de mots et chaque mot une liste
   de trigrames de caractères'''
  f = open(data, "r", encoding="utf-8")
  lines = [[x.rstrip('\n')] for x in f]
  phrases = []
  for phrase in lines:
    temp = phrase[0].split(" ")
    trigram_phrases = []
    for word in temp:
      if word == '<s>' or word == '</s>' or len(word) < 3:
        trigram_phrases.append(word)
        continue
      trigrams = [word] + ["-"+word[max(0, i-2):i+1] for i in range(1, len(word)+1)]
      #trigrams.append(word[-2:])  # Add the end of the word
      
      trigram_phrases.append(trigrams)
    phrases.append(trigram_phrases)
  return phrases

print(separation(data))

# Create lists for words and lexicon
word_list = []
lexicon = []

# Process the separated phrases
phrases = separation(data)
for phrase in phrases:
  sentence_word_list = []
  sentence_lexicon = []
  for word in phrase:
    if isinstance(word, list):  # If the word is a list of trigrams
      original_word = word[0]  # Join trigrams to form the original word
      sentence_word_list.append((original_word, len(word)))
      sentence_lexicon.extend(word[1:])
    else:
      sentence_word_list.append((word, 1))
      sentence_lexicon.append(word)
  word_list.append(sentence_word_list)
  lexicon.append(sentence_lexicon)

print("Word List:", word_list)
print("Lexicon:", lexicon)
print(len(lexicon))

# Create a filtered lexicon
m = 5  # Define the minimum count threshold
word_count = {}

# Count the occurrences of each word
for sentence in word_list:
  for word, count in sentence:
    if word in word_count:
      word_count[word] += 1
    else:
      word_count[word] = 1

# Filter words and their trigrams based on the count threshold
filtered_word_list = []
filtered_lexicon = []

for sentence in word_list:
  filtered_sentence = []
  filtered_trigrams = []
  for word, count in sentence:
    if word_count[word] >= m:
      filtered_sentence.append((word, count))
      for trigram in lexicon[word_list.index(sentence)]:
        filtered_trigrams.append(trigram)
  filtered_word_list.append(filtered_sentence)
  filtered_lexicon.append(filtered_trigrams)



print("Filtered word list:", filtered_word_list)
print("Filtered lexicon:", filtered_lexicon)


```

```python
import torch

def generate_ngrams(word, n):
  ngrams = [word[i:i+1] for i in range(len(word)-n+1)]
  return ngrams



def w2v(txt, n, L, k, eta, e, minc):
  if torch.cuda.is_available():
    device = torch.device('cuda')
    print('GPU available')
    
  else:
    device = torch.device('cpu')

  filtered_lex_count = filtered_word_list
  filtered_lex = filtered_lexicon

  tot_mots_lex = 0
  for i in range(len(filtered_lex_count)):
    tot_mots_lex += filtered_lex_count[i][1]
  tot_mots_lex

  M = torch.randn((len(filtered_lex), n), requires_grad=True, device=device)
  C = torch.randn((len(filtered_lex), n), requires_grad=True, device=device)

  for iter in range(e):
    print('Itération ' + str(iter + 1))
    for mot in range(len(filtered_lex)):
      batch = select_cposcneg(filtered_lex[mot], L, filtered_lex, filtered_lex_count, k, tot_mots_lex)
      mot_plong = M[mot]
      cpos_plong = C[batch[0]]

      cneg_plong = torch.empty((0, n), requires_grad=True, device=device)
      for cneg in batch[1]:
        a = torch.reshape(C[cneg], (1, n))
        cneg_plong = torch.concat((cneg_plong, a), 0)

      mot_plong.retain_grad()
      cpos_plong.retain_grad()
      cneg_plong.retain_grad()

      loss = -(torch.log(torch.nn.functional.sigmoid(cpos_plong.T @ mot_plong)) + torch.sum(torch.log(torch.nn.functional.sigmoid(-cneg_plong @ mot_plong))))

      loss.backward(retain_graph=True)

      mot_plong = mot_plong - eta * (mot_plong.grad)
      cpos_plong = cpos_plong - eta * (cpos_plong.grad)
      cneg_plong = cneg_plong - eta * (cneg_plong.grad)

      ngrams = generate_ngrams(filtered_lex[mot], 3)
      for ngram in ngrams:
        ngram_idx = filtered_lex.index(ngram) if ngram in filtered_lex else -1
        if ngram_idx != -1:
          ngram_plong = M[ngram_idx]
          ngram_plong.retain_grad()
          ngram_plong = ngram_plong - eta * (ngram_plong.grad)
          
      with torch.no_grad():
        M[mot] = mot_plong
        C[batch[0]] = cpos_plong
        for i in range(len(batch[1])):
          C[batch[1][i]] = cneg_plong[i]

  plongements = [str(len(filtered_lex)) + ' ' + str(n)]

  with torch.no_grad():
    for i in range(len(filtered_lex)):
      delimiter = " "  # Define a delimiter
      if torch.cuda.is_available():
        num_list_string = map(str, M.cpu()[i].numpy())  # Convert each element into a string first
      else:
        num_list_string = map(str, M[i].numpy())
      join_num_str = delimiter.join(num_list_string)
      plongements.append(filtered_lex[i] + ' ' + join_num_str)

  with open('plongements_' + 'results' + '.txt', 'w', encoding="utf-8") as f:
    for line in plongements:
      f.write(f"{line}\n")

  return M, C, plongements
```
